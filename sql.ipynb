We walk through the incremental forms of the SQL operators from §4.2 of [the DBSP paper](https://arxiv.org/pdf/2203.16684), demonstrating each one on a small example. Together, the examples below exercise the relational core of the library: union, project, select, join, intersect, except, and the group-by aggregates for sum and max.

Every example follows a shape that recurs throughout the notebook. A one-dimensional `Input` stands in for each source table, while a `LiftStreamIntroduction` lifts the stream onto the two-dimensional body lattice. The view at the latest tick is read via `Evaluator.latest(op)`, which substitutes `0` for any `OMEGA` coordinates in the maximal frontier element. The inner axis stays at zero throughout, because none of the examples below needs a fixpoint to converge.

```python
from pydbsp.circuit import Circuit
from pydbsp.compute import ComputeCtx
from pydbsp.core import Antichain, dbsp_time
from pydbsp.evaluate import Evaluator
from pydbsp.operator import Input, Lift1, LiftStreamIntroduction
from pydbsp.progress import Time
from pydbsp.relational_operators import (
    DeltaLiftedDeltaLiftedDistinct,
    DeltaLiftedDeltaLiftedJoin,
    LiftProject,
    LiftSelect,
)
from pydbsp.operator import Integrate
from pydbsp.storage import DictStorage
from pydbsp.zset import ZSet, ZSetAddition


def make_eval(group):
    return Evaluator[Time](
        circuit=Circuit[Time](),
        storage=DictStorage(),
        ctx=ComputeCtx(lattice=dbsp_time(2)),
        group=group,
    )


def make_input(e, lattice_nestedness=1):
    return Input(frontier=Antichain(dbsp_time(lattice_nestedness))).connect(e.circuit, ())


# (SELECT * FROM I1) UNION (SELECT * FROM I2)
g = ZSetAddition[int]()
e = make_eval(g)
a = make_input(e)
b = make_input(e)
a_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (a,))
b_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (b,))
# ``added`` is the pointwise Z-set sum of the two operands.
from pydbsp.operator import Lift2
added = Lift2[ZSet[int], ZSet[int], ZSet[int]](op=g.add).connect(e.circuit, (a_2d, b_2d))
# ``distinct`` is Δ-in Δ-out: emits the *change* to the distinct
# view at each outer tick.
distinct = DeltaLiftedDeltaLiftedDistinct[int](inner_group=g).connect(e.circuit, (added,))
# ``union`` integrates ``distinct`` on the outer axis, turning the
# Δ-stream back into the cumulative SQL view.
union = Integrate[ZSet[int]](group=g).connect(e.circuit, (distinct,))

e.push(a, ZSet({1: 1}))
e.push(a, ZSet({2: 1}))
e.push(a, ZSet({3: -1}))
print(e.latest(distinct))
print(e.latest(union))
e.push(b, ZSet({2: 1}))
e.push(b, ZSet({3: 1}))
print(e.latest(distinct))
print(e.latest(union))
e.push(b, ZSet({}))
print(e.latest(distinct))
print(e.latest(union))
```

```python
# SELECT DISTINCT I.c[0] FROM I
g_in = ZSetAddition[tuple[str, str]]()
g_out = ZSetAddition[str]()

e = make_eval(g_in)
i_in = make_input(e)
i_2d = LiftStreamIntroduction[ZSet[tuple[str, str]]](group=g_in).connect(e.circuit, (i_in,))
proj = LiftProject[tuple[str, str], str](f=lambda x: x[0]).connect(e.circuit, (i_2d,))
projection = DeltaLiftedDeltaLiftedDistinct[str](inner_group=g_out).connect(e.circuit, (proj,))

e.push(i_in, ZSet({("a", "b"): 1}))
print(e.latest(projection))
e.push(i_in, ZSet({("c", "d"): 1}))
print(e.latest(projection))
e.push(i_in, ZSet({("e", "f"): 1}))
print(e.latest(projection))
```

```python
# SELECT DISTINCT * FROM I WHERE p(I.c)
g = ZSetAddition[tuple[str, str]]()
e = make_eval(g)
i_in = make_input(e)
i_2d = LiftStreamIntroduction[ZSet[tuple[str, str]]](group=g).connect(e.circuit, (i_in,))
sel = LiftSelect[tuple[str, str]](pred=lambda x: x[0] == "a").connect(e.circuit, (i_2d,))
filtering = DeltaLiftedDeltaLiftedDistinct[tuple[str, str]](inner_group=g).connect(e.circuit, (sel,))

e.push(i_in, ZSet({("a", "b"): 1}))
print(e.latest(filtering))
e.push(i_in, ZSet({("a", "d"): 1}))
print(e.latest(filtering))
e.push(i_in, ZSet({("e", "f"): 1}))
print(e.latest(filtering))
```

```python
# SELECT DISTINCT f(I.c) FROM I — projection by an arbitrary function
g_in = ZSetAddition[tuple[int, str]]()
g_out = ZSetAddition[int]()

e = make_eval(g_in)
i_in = make_input(e)
i_2d = LiftStreamIntroduction[ZSet[tuple[int, str]]](group=g_in).connect(e.circuit, (i_in,))
proj = LiftProject[tuple[int, str], int](f=lambda x: x[0] ** 2).connect(e.circuit, (i_2d,))
selection = DeltaLiftedDeltaLiftedDistinct[int](inner_group=g_out).connect(e.circuit, (proj,))

e.push(i_in, ZSet({(1, "b"): 1}))
print(e.latest(selection))
e.push(i_in, ZSet({(2, "d"): 1}))
print(e.latest(selection))
e.push(i_in, ZSet({(3, "f"): 1}))
print(e.latest(selection))
```

```python
# SELECT I1.*, I2.* FROM I1, I2 — pred always true = full cartesian
g_in = ZSetAddition[int]()
g_out = ZSetAddition[tuple[int, int]]()

e = make_eval(g_in)
a_in = make_input(e)
b_in = make_input(e)
a_2d = LiftStreamIntroduction[ZSet[int]](group=g_in).connect(e.circuit, (a_in,))
b_2d = LiftStreamIntroduction[ZSet[int]](group=g_in).connect(e.circuit, (b_in,))
cartesian = DeltaLiftedDeltaLiftedJoin[int, int, tuple[int, int]](
    pred=lambda l, r: True,
    proj=lambda l, r: (l, r),
    group_a=g_in,
    group_b=g_in,
    out_group=g_out,
).connect(e.circuit, (a_2d, b_2d))

e.push(a_in, ZSet({1: 1}))
e.push(b_in, ZSet({1: 1}))
print(e.latest(cartesian))
e.push(a_in, ZSet({2: 1}))
e.push(b_in, ZSet({2: 1}))
print(e.latest(cartesian))
e.push(a_in, ZSet({3: 1}))
e.push(b_in, ZSet({3: 1}))
print(e.latest(cartesian))
```

```python
# SELECT I1.*, I2.* FROM I1 JOIN I2 ON I1.c1 = I2.c2
g_in = ZSetAddition[tuple[str, str]]()
g_out = ZSetAddition[tuple[tuple[str, str], tuple[str, str]]]()

e = make_eval(g_in)
a_in = make_input(e)
b_in = make_input(e)
a_2d = LiftStreamIntroduction[ZSet[tuple[str, str]]](group=g_in).connect(e.circuit, (a_in,))
b_2d = LiftStreamIntroduction[ZSet[tuple[str, str]]](group=g_in).connect(e.circuit, (b_in,))
join_ = DeltaLiftedDeltaLiftedJoin[tuple[str, str], tuple[str, str], tuple[tuple[str, str], tuple[str, str]]](
    pred=lambda l, r: l[0] == r[0],
    proj=lambda l, r: (l, r),
    group_a=g_in,
    group_b=g_in,
    out_group=g_out,
).connect(e.circuit, (a_2d, b_2d))

e.push(a_in, ZSet({("a", "b"): 1}))
e.push(b_in, ZSet({("a", "b"): 1}))
print(e.latest(join_))
e.push(a_in, ZSet({("a", "d"): 1}))
e.push(b_in, ZSet({("c", "d"): 1}))
print(e.latest(join_))
e.push(a_in, ZSet({("e", "f"): 1}))
e.push(b_in, ZSet({("e", "g"): 1}))
print(e.latest(join_))
```

```python
# (SELECT * FROM I1) INTERSECT (SELECT * FROM I2)
g = ZSetAddition[int]()
e = make_eval(g)
a_in = make_input(e)
b_in = make_input(e)
a_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (a_in,))
b_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (b_in,))
intersection = DeltaLiftedDeltaLiftedJoin[int, int, int](
    pred=lambda l, r: l == r,
    proj=lambda l, r: l,
    group_a=g, group_b=g, out_group=g,
).connect(e.circuit, (a_2d, b_2d))

e.push(a_in, ZSet({1: 1}))
e.push(b_in, ZSet({1: 1}))
print(e.latest(intersection))
e.push(a_in, ZSet({2: 1}))
e.push(b_in, ZSet({2: 1}))
print(e.latest(intersection))
e.push(a_in, ZSet({3: 1}))
e.push(b_in, ZSet({3: -1}))
print(e.latest(intersection))
```

```python
# SELECT * FROM I1 EXCEPT SELECT * FROM I2 — (a − b) with incremental distinct
from pydbsp.operator import Lift2

g = ZSetAddition[int]()
e = make_eval(g)
a_in = make_input(e)
b_in = make_input(e)
a_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (a_in,))
b_2d = LiftStreamIntroduction[ZSet[int]](group=g).connect(e.circuit, (b_in,))
neg_b = Lift1[ZSet[int], ZSet[int]](f=g.neg).connect(e.circuit, (b_2d,))
diff_ = Lift2[ZSet[int], ZSet[int], ZSet[int]](op=g.add).connect(e.circuit, (a_2d, neg_b))
difference = DeltaLiftedDeltaLiftedDistinct[int](inner_group=g).connect(e.circuit, (diff_,))

e.push(a_in, ZSet({1: 1}))
e.push(b_in, ZSet({1: 1}))
print(e.latest(difference))
e.push(a_in, ZSet({2: 1}))
e.push(b_in, ZSet({2: 1}))
print(e.latest(difference))
e.push(a_in, ZSet({3: 1}))
e.push(b_in, ZSet({4: 1}))
print(e.latest(difference))
```

```python
from pydbsp.operator import Differentiate, Integrate

# SELECT I.k, SUM(I.v) FROM I GROUP BY I.k
#
# DBSP's ``aggregate`` pattern: integrate the stream of diffs to get
# the cumulative ZSet at each tick, apply the (non-linear) aggregation
# as a pointwise function, then differentiate back to a diff stream.
# No fixpoint here, so everything fits on a 1-D lattice.


def zset_group_sum(z):
    print(f"zset_group_sum: {z}")
    totals: dict = {}
    for (k, v), w in z.inner.items():
        totals[k] = totals.get(k, 0) + v * w
    return ZSet({(k, v): 1 for k, v in totals.items()})


g_in = ZSetAddition[tuple[str, int]]()
g_out = ZSetAddition[tuple[str, int]]()

e = Evaluator[Time](
    circuit=Circuit[Time](),
    storage=DictStorage(),
    ctx=ComputeCtx(lattice=dbsp_time(1)),
    group=g_in,
)
i_in = Input[ZSet[tuple[str, int]]](frontier=Antichain(dbsp_time(1))).connect(e.circuit, ())
cum = Integrate[ZSet[tuple[str, int]]](group=g_in).connect(e.circuit, (i_in,))
agg = Lift1[ZSet[tuple[str, int]], ZSet[tuple[str, int]]](f=zset_group_sum).connect(e.circuit, (cum,))
agg_diffs = Differentiate[ZSet[tuple[str, int]]](group=g_out).connect(e.circuit, (agg,))

e.push(i_in, ZSet({("a", 2): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("a", 3): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("b", 1): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("a", 5): -1}))
print(f"latest: {e.latest(agg_diffs)}")
```

```python
from pydbsp.indexed_relational_operators import (
    DeltaLiftedDeltaLiftedGroupBy,
    LiftLiftIndex,
)
from pydbsp.indexed_zset import IndexedZSetAddition

from pydbsp.operator import Differentiate, Integrate

# SELECT I.k, SUM(I.v) FROM I GROUP BY I.k
#
# DBSP's ``aggregate`` pattern: integrate the stream of diffs to get
# the cumulative ZSet at each tick, apply the (non-linear) aggregation
# as a pointwise function, then differentiate back to a diff stream.
# No fixpoint here, so everything fits on a 1-D lattice.

def zset_group_sum(z):
    print(f"zset_group_sum: {z}")
    totals: dict = {}
    for (k, v), w in z.inner.items():
        totals[k] = totals.get(k, 0) + v * w
    return ZSet({(k, v): 1 for k, v in totals.items()})

g_in = ZSetAddition[tuple[str, int]]()
g_out = ZSetAddition[tuple[str, int]]()
g_idx = IndexedZSetAddition[str, str](g_in, lambda w: w[0])

e = Evaluator[Time](
    circuit=Circuit[Time](),
    storage=DictStorage(),
    ctx=ComputeCtx(lattice=dbsp_time(2)),
    group=g_in,
)

i_in = Input[ZSet[tuple[str, int]]](frontier=Antichain(dbsp_time(1))).connect(e.circuit, ())
words = Lift1[ZSet[str], ZSet[str]](f=zset_group_sum).connect(e.circuit, (i_in,))
words_2d = LiftStreamIntroduction[ZSet[str]](group=g_in).connect(e.circuit, (words,))
indexed = LiftLiftIndex[str, str](indexer=lambda w: w[0]).connect(e.circuit, (words_2d,))
ktable = DeltaLiftedDeltaLiftedGroupBy[str, str, int](
    aggregate=lambda items: sum([x[1] for x, w in items]),
    group=g_idx,
    out_group=g_out,
).connect(e.circuit, (indexed,))
agg_diffs = Integrate[ZSet[tuple[str, int]]](group=g_out).connect(e.circuit, (ktable,))
# cum = Integrate[ZSet[tuple[str, int]]](group=g_in).connect(e.circuit, (i_in,))
# agg = Lift1[ZSet[tuple[str, int]], ZSet[tuple[str, int]]](f=zset_group_sum).connect(e.circuit, (cum,))
# agg_diffs = Differentiate[ZSet[tuple[str, int]]](group=g_out).connect(e.circuit, (agg,))

e.push(i_in, ZSet({("a", 2): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("a", 3): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("b", 1): 1}))
print(f"latest: {e.latest(agg_diffs)}")
e.push(i_in, ZSet({("a", 5): -1}))
print(f"latest: {e.latest(agg_diffs)}")
```

```python
# SELECT I.k, MAX(I.v) FROM I GROUP BY I.k
#
# Same pattern as SUM — a different pointwise aggregation function.
# Because MAX is non-linear, a retraction (negative weight) on the
# max-holder properly exposes the new max.


def zset_group_max(z):
    positive_by_key: dict = {}
    for (k, v), w in z.inner.items():
        if w <= 0:
            continue
        cur = positive_by_key.get(k)
        if cur is None or v > cur:
            positive_by_key[k] = v
    return ZSet({(k, v): 1 for k, v in positive_by_key.items()})


g_in = ZSetAddition[tuple[str, int]]()
g_out = ZSetAddition[tuple[str, int]]()

e = Evaluator[Time](
    circuit=Circuit[Time](),
    storage=DictStorage(),
    ctx=ComputeCtx(lattice=dbsp_time(1)),
    group=g_in,
)
i_in = Input[ZSet[tuple[str, int]]](frontier=Antichain(dbsp_time(1))).connect(e.circuit, ())
cum = Integrate[ZSet[tuple[str, int]]](group=g_in).connect(e.circuit, (i_in,))
max_agg = Lift1[ZSet[tuple[str, int]], ZSet[tuple[str, int]]](f=zset_group_max).connect(e.circuit, (cum,))
max_diffs = Differentiate[ZSet[tuple[str, int]]](group=g_out).connect(e.circuit, (max_agg,))

e.push(i_in, ZSet({("a", 2): 1}))
print(e.latest(max_diffs))
e.push(i_in, ZSet({("a", 3): 1}))
print(e.latest(max_diffs))
e.push(i_in, ZSet({("b", 1): 1}))
print(e.latest(max_diffs))
e.push(i_in, ZSet({("a", 3): -1}))
print(e.latest(max_diffs))
```
